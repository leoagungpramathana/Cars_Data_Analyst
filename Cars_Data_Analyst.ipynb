**Q) 1. Data Cleansing** ✅

Find all Null, Nan Value in the Dataset. If there is any Null value in any column, then fill it with the mean of that column.

```python
car.isnull()
```

```python
car.isnull().sum()
```

```python
car.dtypes
```



*   Dikarenakan Nan ada di kolom "Make" dengan type data "Object", kita bisa mengisi kolom Nan yang kosong dengan "Unknown_Make" dan tidak diperlukan mencari mean dari kolom "Make" tersebut.



```python
print("Unique value awal:\n", car["Make"].unique())
```

```python
car["Make"] = car["Make"].fillna("Unknown_Make")
```

```python
print("Unique value setelah fillna:\n", car["Make"].unique())
```

```python
print("Unique value awal:\n", car["Model"].unique())
```

```python
car["Model"] = car["Model"].fillna("Unknown_Model")
```

```python
print("Unique value setelah fillna:\n", car["Model"].unique())
```

```python
print("Unique value awal Type:\n", car["Type"].unique())
car["Type"] = car["Type"].fillna("Unknown_Type")
print("Unique value setelah fillna Type:\n", car["Type"].unique())

print("Unique value awal Origin:\n", car["Origin"].unique())
car["Origin"] = car["Origin"].fillna("Unknown_Origin")
print("Unique value setelah fillna Origin:\n", car["Origin"].unique())

print("Unique value awal DriveTrain:\n", car["DriveTrain"].unique())
car["DriveTrain"] = car["DriveTrain"].fillna("Unknown_DriveTrain")
print("Unique value setelah fillna DriveTrain:\n", car["DriveTrain"].unique())
```



*   Nah, untuk kolom "MSRP" dan "Invoice", dikarenakan kolom ini type data nya "Object" dan value nya adalah hitungan Dolar, maka kita tidak harus membuat mean dari kolom tersebut.
Jika ingin mengisi Nan dari kolom "MSRP" dan "Invoice", maka diharuskan mengganti type data "float64" dahulu baru bisa mencari mean nya.
Dan di kasus ini juga kita tidak mencari harga dari mobil-mobil tersebut, kita tidak perlu merubah apapun.
Atau mungkin bisa kita "dropna" saja karena kolom tersebut tidak kita gunakan.



```python
print("Unique value awal MSRP:\n", car["MSRP"].unique())
print("Unique value awal Invoice:\n", car["Invoice"].unique())
```

```python
car.drop('MSRP', axis=1, inplace=True)
car.drop('Invoice', axis=1, inplace=True)
```



*   Menambahkan field kosong Nan dengan mean, dikarenakan kolom ini ber type data float64 dengan value angka, jadi bisa kita cari rata" nya dari value tersebut.



```python
car['Cylinders'].fillna(car['Cylinders'].mean(), inplace = True)
```

```python
car['EngineSize'].fillna(car['EngineSize'].mean(), inplace = True)
```

```python
car['Horsepower'].fillna(car['Horsepower'].mean(), inplace = True)
```

```python
car['MPG_City'].fillna(car['MPG_City'].mean(), inplace = True)
```

```python
car['MPG_Highway'].fillna(car['MPG_Highway'].mean(), inplace = True)
car['Weight'].fillna(car['Weight'].mean(), inplace = True)
car['Wheelbase'].fillna(car['Wheelbase'].mean(), inplace = True)
car['Length'].fillna(car['Length'].mean(), inplace = True)
```

**Q) 4. Instruction (Removing unwanted records) ✅**



*   Remove all the records (row) where Weight is above 4000



```python
car.head(2)
```

```python
car[car['Weight'] > 4000]
```



*   Tanda tilde ( ~ ) di Pandas DataFrame adalah operator logika yang digunakan untuk membalikkan larik boolean. Dengan kata lain, operator ini mengambil nilai boolean dan mengembalikan kebalikannya. Sebagai contoh, jika Anda memiliki larik boolean yang bernilai True , tanda tilde akan mengembalikan False



```python
car[~(car['Weight'] > 4000)]
```
